```python
import numpy as np

tokens = list(range(1_000_000))
print(tokens[:10])
print(tokens[-5:])
m1=30/272000
m2=(45-35)/(1000000-272000)
precio = []
for t in tokens:
    if t <= 272_000:
        precio.append(30* m1 + t)  # MAL: reemplazar con la fórmula correcta
    else:
        precio.append(45 * m2 + (t - 272000))  # MAL: reemplazar con la fórmula correcta
import matplotlib.pyplot as plt

plt.plot(tokens, precio)
plt.xlabel("Tokens")
plt.ylabel("Precio")
plt.title("Precio vs Tokens")
plt.show()
```

```python
# Calculate intercepts for continuity at t=272,000
b1 = 30
b2 = b1 + m1 * 272_000 - m2 * 272_000

precio = []
for t in tokens:
    if t <= 272_000:
        precio.append(m1 * t + b1)
    else:
        precio.append(m2 * t + b2)

plt.figure(figsize=(10, 6))
plt.plot(tokens, precio, linewidth=2)
plt.xlabel("Tokens")
plt.ylabel("Precio")
plt.title("Precio vs Tokens (Piecewise Linear)")
plt.grid(True, alpha=0.3)
plt.show()
```
