```python
import numpy as np
import matplotlib.pyplot as plt

# sistema Ax = B
A = np.array([
    [-5.27, 15],
    [114.514, -5.27]
])

B = np.array([
    15.14,
    -65.53
])

# resolver
m, b = np.linalg.solve(A, B)

print("m =", m)
print("b =", b)

# tus puntos
xs = np.array([
    3.38, 0.35, 2.07, -0.45, -0.55,
    -1.06, -2.77, 3.08, -3.98, -5.0,
    -3.18, -1.96, 2.37, -1.66, 4.09
])

ys = np.array([
    -0.04, 1.28, 0.65, 2.18, 1.70,
    1.96, 3.36, 0.18, 3.35, 4.16,
    2.95, 2.34, 0.38, 1.75, -1.06
])

# recta
x_line = np.linspace(min(xs), max(xs), 200)
y_line = m*x_line + b

# gráfica
plt.figure(figsize=(8,5))
plt.scatter(xs, ys, color='blue', label='Datos')
plt.plot(x_line, y_line, color='red', label=f'y = {m:.3f}x + {b:.3f}')

plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.array([
    3.38, 0.35, 2.07, -0.45, -0.55,
    -1.06, -2.77, 3.08, -3.98, -5.0,
    -3.18, -1.96, 2.37, -1.66, 4.09
])

ys = np.array([
    -0.04, 1.28, 0.65, 2.18, 1.70,
    1.96, 3.36, 0.18, 3.35, 4.16,
    2.95, 2.34, 0.38, 1.75, -1.06
])


m, b = np.polyfit(xs, ys, 1)
print(m,b)

a, b2, c = np.polyfit(xs, ys, 2)
print(a,b2,c)

x_line = np.linspace(min(xs), max(xs), 300)

y_linear = m * x_line + b
y_quad = a * x_line**2 + b2 * x_line + c


plt.figure(figsize=(8,5))

plt.scatter(xs, ys, color="black", label="Datos")

plt.plot(x_line, y_linear, color="blue", label="Lineal")
plt.plot(x_line, y_quad, color="red", label="Cuadrática")

plt.grid(True)
plt.legend()
plt.title("lineal vs cuadrática")

plt.show()
```

A partir de esta grafica se puede observar que realmente el coeficiente que acompaña al coeficiente a es despreciable y por tanto se puede asumir que la linea cumple adecuadamente con el objetivo de la practica.

```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.array([
    3.38, 0.35, 2.07, -0.45, -0.56,
    -1.06, -2.78, 3.08,
    -3.99, -5.00, -3.18, -1.97,
    2.37, -1.67, 4.09,
    -4.60, 2.68, 2.78, -3.79, -1.87
])

ys = np.array([
    15.65, -11.20, -4.00, 18.03, 7.94,
    15.32, -4.40, 1.39,
    -11.92, -90.24, 6.92, 28.35,
    -2.41, -5.47, 35.91,
    -55.53, 0.77, 4.79, -27.05, 2.85
])

titulos = ['Lineal', 'Cuadrática', 'Cúbica']
colores_curva = ['royalblue', 'tomato', 'darkorange']

for grado, titulo, color_curva in zip([1, 2, 3], titulos, colores_curva):
    coefs = np.polyfit(xs, ys, grado)
    y_pred = np.polyval(coefs, xs)
    errores = np.abs(ys - y_pred)

    error_promedio = errores.mean()
    desv_std = errores.std()
    idx_max = np.argmax(errores)
    idx_min = np.argmin(errores)

    x_line = np.linspace(min(xs) - 0.3, max(xs) + 0.3, 400)
    y_line = np.polyval(coefs, x_line)

    fig, ax = plt.subplots(figsize=(10, 7))

    ax.plot(x_line, y_line, color=color_curva, linewidth=2, label=f'Regresión {titulo}')

    for i in range(len(xs)):
        ax.plot([xs[i], xs[i]], [ys[i], y_pred[i]], color='gray', linewidth=0.7, linestyle='--', zorder=2)
        offset_x = 0.08
        offset_y = (ys[i] - y_pred[i]) / 2
        ax.annotate(f'{errores[i]:.1f}',
                    xy=(xs[i] + offset_x, y_pred[i] + offset_y),
                    fontsize=7, color='dimgray', va='center')

    ax.scatter(xs, ys, color='steelblue', zorder=4, s=60, label='Datos')
    ax.scatter(xs, y_pred, color=color_curva, zorder=3, s=35, marker='x')

    ax.scatter(xs[idx_max], ys[idx_max], color='red', s=180, zorder=5,
               label=f'Error máx: {errores[idx_max]:.2f}  (x={xs[idx_max]:.2f})')
    ax.scatter(xs[idx_min], ys[idx_min], color='limegreen', s=180, zorder=5,
               label=f'Error mín: {errores[idx_min]:.2f}  (x={xs[idx_min]:.2f})')

    ax.set_title(f'Regresión {titulo}', fontsize=13, fontweight='bold')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True, alpha=0.4)
    ax.legend(fontsize=9, loc='upper left')
    ax.text(0.02, 0.01,
            f'Error promedio: {error_promedio:.4f}\nDesv. estándar: {desv_std:.4f}',
            transform=ax.transAxes, fontsize=9,
            verticalalignment='bottom', horizontalalignment='left',
            bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.show()
```
