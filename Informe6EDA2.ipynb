### Ejercicio 1 
---

```python
def dfs(nodo, grafo, searched, componentR, arbolDFS):

    componentR.append(nodo)   # agregar nodo al recorrido
    searched[nodo] = True     # marcar como visitado

    print('Vecinos de', nodo, ':', grafo[nodo])

    # recorrer vecinos
    for vecino in grafo[nodo]:

        # si no fue visitado
        if not searched[vecino]:

            # agregar arista al árbol DFS
            arbolDFS.append((nodo, vecino))

            # llamada recursiva
            dfs(vecino, grafo, searched, componentR, arbolDFS)

            print('Finaliza', vecino)
            print('Vuelve a', nodo)
            print()


grafo = {
    'Start': ['A', 'B', 'C'],
    'A': ['G'],
    'B': ['C', 'E'],
    'C': ['D'],
    'D': ['F'],
    'E': ['D', 'F'],
    'F': ['D'],
    'G': ['E']
}

nodoS = 'Start'

# diccionario de visitados
searched = {}

# inicialmente todos en False
for nodo in grafo:
    searched[nodo] = False

arbolDFS = []
componentR = []

dfs(nodoS, grafo, searched, componentR, arbolDFS)

print("La búsqueda DFS es:", componentR)
print()

print("El árbol DFS es (Padre -> Hijo):")
print(arbolDFS)
```
