```python
import numpy as np

xs = list(np.linspace(-2,4,50))

def f(x):
    return x**3 -3*x**2+x-1

yx = [f(x) for x in xs]
```

```python
# Graficar xs vs yx, marcar la raíz y mostrar líneas punteadas
import matplotlib.pyplot as plt

plt.plot(xs, yx, "o:")
plt.axvline(raiz, color='red', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title("f(x) y raíz")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton


historialX = []

def funcionObjetivo(x):
    yValor = x**3 - 3*x**2 + x - 1
    historialX.append(x)
    return yValor

def derivadaFuncion(x):
    yPrimeraDerivada = 3*x**2 - 6*x + 1
    return yPrimeraDerivada


raizEncontrada = newton(funcionObjetivo, x0=100, fprime=derivadaFuncion)


valoresX = np.linspace(2, 5, 200)
valoresY = [val**3 - 3*val**2 + val - 1 for val in valoresX]


plt.figure(figsize=(10, 6))
plt.plot(valoresX, valoresY, color="blue", linewidth=2, label="f(x)")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.axvline(raizEncontrada, color='red', linestyle='-', linewidth=2, label=f"Raíz: {raizEncontrada:.4f}")


for indice, valorX in enumerate(historialX):
    if 2 <= valorX <= 5:
        plt.axvline(valorX, color='green', linestyle=':', alpha=0.7)
        imagenY = valorX**3 - 3*valorX**2 + valorX - 1
        plt.plot(valorX, imagenY, marker='o', color='green', markersize=5)
        plt.text(valorX, 1, f"x{indice}", rotation=90, color='green', verticalalignment='bottom')


plt.title("Iteraciones del Método de Newton sobre f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim(2.5, 5)
plt.ylim(-1, 50)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
##Estimaciones Iniciales 
estimacionesIniciales = [100, 50, 10, 3, -5]

for aproximacionInicial in estimacionesIniciales:
    historialX.clear() 
    raizProgreso = newton(funcionObjetivo, x0=aproximacionInicial, fprime=derivadaFuncion)
    print(f"Con x0 = {aproximacionInicial:<4} -> Se necesitaron {len(historialX)} iteraciones. Raíz: {raizProgreso:.4f}")
```
